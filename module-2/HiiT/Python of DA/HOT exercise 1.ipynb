```python
# Pull out the information of all individuals above the age of 30

db.style.highlight_max('Age') # i did this to know the oldest person

db.style.highlight_between('Age', left=31, color='grey')

# then use conditional formatting to highlight the maximum SALARY and the maximum AGE 
db.style.highlight_max('Salary', color='green').highlight_max('Age', color='brown')

#all together
db.style.highlight_max('Salary', color='Yellow').highlight_between('Age', left=31, color='grey').highlight_max('Age', color='blue')
```
