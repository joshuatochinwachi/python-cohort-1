### IF statement (replica of CASE statement in SQL and IF() / IFS() in Excel)

```python
def age (value):
    if value > 30:
        return 'Pass'
    else:
        return 'Fail'
```

```python
db['Grade'] = db['Age'].apply(age)
```

```python
db
```

```python
def scores (value):
    if value >= 30:
        return 'A'
    elif value >= 20:
        return 'B'
    elif value >= 10:
        return 'C'
    else:
        return 'Fail'
```

```python
db['Result'] = db['Age'].apply(scores)
```

```python
db
```
