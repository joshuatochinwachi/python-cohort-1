## Seaborn

Seaborn is a super easy tool in Python that helps you draw pretty pictures (called plots) of your data. It’s like a magic crayon box built on top of Matplotlib, making graphs look nice and simple to create for beginners! 

**Why Use It?**  
It turns numbers into cool, colorful pictures so you can see patterns fast.

```python
# !pip install seaborn
```

```python
# pip freeze > requirements.txt
```

```python
import seaborn as sns
import pandas as pd
```

### Basic setup

```python
student = ['Daniel', 'Alfred', 'Amah', 'Divine', 'Ruth', 'George']
scores = [80, 90, 70, 60, 85, 74]

data = pd.DataFrame({'Student' : student, 'Scores': scores})
print(data)
```

### Line plot

```python
sns.lineplot(x='Student', y='Scores', data=data, color='blue')
plt.title('Student Test Scores')
plt.show()
```

### Scatter plot

```python
sns.scatterplot(x='Student', y='Scores', data = data, color = 'red')

plt.title('Student test scores')
plt.show()
```

### Bar plot

```python
sns.barplot(x='Student', y='Scores', data=data, color = 'green')

plt.title('Student Test Score')
plt.show()
```

### Histogram

```python
sns.histplot(data=data, x='Scores', color = 'purple')

plt.title('How scores pile up')
plt.show()

# if u save as .png, it'll have a transparent background that you can add to any background for better customizations
```

## Visualizing Web3 Token Transfer Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\Jo$h\Desktop\Visual Studio Code\python-tutorial-cohort-1\module-2\web3_token_transfers.csv', parse_dates = ['timestamp'])
df['date'] = df['timestamp'].dt.date

# Setting Seaborn theme
sns.set(style='whitegrid')

df.head()
```

### Token Transfer Volume Distribution

```python
plt.figure(figsize=(10, 6))

sns.histplot(df['amount'], bins=50, kde=True, color='skyblue')  # Kernel Density Estimate (KDE)
plt.title("Distribution of Token Transfer Amounts")
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.xlim(0, 3000)
plt.show()
```

### Daily Total transfer

```python
daily_volume = df.groupby('date')['amount'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_volume, x='date', y='amount', marker = 'o')
plt.title('Daily Total Token Transfer Volume')
plt.xlabel('Date')
plt.ylabel('Total Amount Transferred')
plt.tight_layout()
plt.show()
```
