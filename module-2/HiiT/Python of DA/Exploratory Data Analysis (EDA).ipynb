### filtering using the query method

```python
#how many autos in the data has a horsepower greater than 100?

df.query('horsepower > 100')
df.query('horsepower > 100').shape
df.query('horsepower > 100').shape[0]


#what are the engine types peculiar to the toyota automobile?
df.query('company == "toyota"')
#or
df.query("company == 'toyota'")
# df.query("company == 'toyota'")['price'].sum()
df.query("company == 'toyota'")['engine_type']
df.query("company == 'toyota'")['engine_type'].value_counts()
```

```python
df.query('    company == "toyota"   &  horsepower > 100        ')
```

```python
df.query('    company == "toyota"   &  horsepower > 100        ').loc[54]
```

#### & (and),  | (or)

### Filtering using the .loc[] method

```python
df['company'] == 'toyota'

df.loc[df['company'] == 'toyota']

df.loc[df['company'] == 'toyota']['price'].mean()
```

```python
df['horsepower']>200

df.loc[df['horsepower']>200]   #locating rows with hp > 100
```

```python
df.loc[df.duplicated()]   #locating rows where there are duplicates
```

### Filtering with Wildcards

```python
df['company'].unique()
df['company'].value_counts()

#can allow u to deduce the data
```

```python
df['company'].str.startswith('t')
df.loc[df['company'].str.startswith('t')]

df['company'].str.endswith('w')
df.loc[df['company'].str.endswith('w')]

df.loc[df['company'].str.contains('ota')]
df.loc[df['company'].str.contains('zd')]

df.loc[df['company'].str.contains('^t')]  #where it begins with t

df.loc[df['company'].str.contains('^[tb]')]  #where it begins with t or b

df.loc[df['company'].str.contains('^[tT]')]  #where it begins with t or T

df.loc[df['company'].str.contains('^[a-c]')]  #where it begins with a to c

#you do not necessarily need to put the .loc[]. You'll get same result

df[df['company'].str.contains('a$')]  #where it ends with a
```

##### using OR

```python
df.query('  company == "toyota" |  company == "mazda"  ')

df.query('  company == "toyota" |  wheel_base > 100  ')
```

```python
df.query('  company == "toyota" or  wheel_base > 100  ')
```

```python
df.query('  company == "toyota" and  wheel_base > 100  ')
```
