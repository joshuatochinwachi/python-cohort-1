### plot a bar chart showing the numbers of all non-toyota autos in the data

```python
df.query('company != "toyota" ')
df.query('company != "toyota" ').shape
```

```python
df.query('company != "toyota" ')[['company', 'price']].groupby('company').count()
```

```python
df.query('company != "toyota" ')[['company', 'price']].groupby('company').count().plot(kind='bar')
```

```python
df.query('company != "toyota" ')[['company', 
                                  'price']].groupby('company').count().plot(kind='pie',y='price')
```

```python
df.query('company != "toyota" ')[['company', 
                                  'price']].groupby('company').count().plot(kind='pie',y='price')
plt.title('NUMBERS ON NON TOYOTA AUTOS')
```

```python
df.query('company != "toyota" ')[['company', 
                                  'price']].groupby('company').count().plot(kind='pie',y='price',
                                                                           legend = False,
                                                                           figsize = (12, 8))
plt.title('NUMBERS ON NON TOYOTA AUTOS')
```
