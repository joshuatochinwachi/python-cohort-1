### Dropping a column/index

```python
df.drop(columns = 'Age')   #dropping 1 column temporary

df.drop(columns = ['Age', 'Tax'])  #dropping 2 column temporary

# df.drop(columns = 'Age',inplace=True)   #making it permanent
```

```python
df.style.highlight_max()
```

```python
df.style.highlight_max('Age')
```

```python
df.style.highlight_max(['Age','Tax'])
```

```python
df.style.highlight_max('Age').highlight_min('Age',color='blue')
df.style.highlight_max('Age').highlight_min('Age',color='blue').to_excel('sample2.xlsx') # to excel
```

```python
df.style.highlight_between()
df.style.highlight_between('Age')
df.style.highlight_between('Age',left=35) #highlighting 35 and above
```

```python
df.style.highlight_between('Age',right=38) #highlighting 38 below
```

###### note: left is min, right is max

```python
df.style.highlight_between('Age',left=25,right=38)   #similar to AND fxn in SQL 
#highlighting btw 25 and 38
```
